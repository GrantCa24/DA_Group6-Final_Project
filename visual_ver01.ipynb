```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
netflix = pd.read_csv('netflix.csv')

# Movie vs. TV Show Distribution
plt.figure(figsize=(10, 6))
netflix['type'].value_counts().plot(kind='bar')
plt.title('Distribution of Movies vs. TV Shows')
plt.xlabel('Type')
plt.ylabel('Number of Titles')
plt.xticks(rotation=0)
plt.show()
```

```python
# Top Countries Producing Content
plt.figure(figsize=(10, 6))
netflix['country'].value_counts().head(10).plot(kind='pie', autopct='%1.1f%%')
plt.title('Top Countries Producing Content on Netflix')
plt.ylabel('')
plt.show()
```

```python
# Rating Distribution
plt.figure(figsize=(10, 6))
sns.histplot(netflix['rating'], kde=True, bins=20)
plt.title('Distribution of Ratings on Netflix')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Duration Analysis
plt.figure(figsize=(10, 6))
sns.boxplot(x='type', y='duration', data=netflix)
plt.title('Duration of Movies and TV Shows')
plt.xlabel('Type')
plt.ylabel('Duration (minutes)')
plt.show()
```
